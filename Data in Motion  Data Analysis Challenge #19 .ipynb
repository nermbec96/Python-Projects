## Q7 What is the average speed? Also display a plot to show it’s distribution.

## Q8 Explore the feature relationships. Are there any positively or negatively correlated relationships?

```python
correlation_matrix = df.corr(method = 'pearson')
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(correlation_matrix,annot = True,linewidth=2, linecolor='black',
            cmap=sns.diverging_palette(200,75, center="dark", as_cmap=True))
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
corr_pairs = correlation_matrix.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
height_sorted_pairs = sorted_pairs[(sorted_pairs) > 0.5]
height_sorted_pairs
```

```python
low_sorted_pairs = sorted_pairs[(sorted_pairs) < 0.3]
low_sorted_pairs.tail(40)
```

## Most common types of materials used to make roller coaster

```python
materials = df.groupby(df['Type']).size()
materials = materials.sort_values(ascending = False).head(5)
print('Most commen type materials used to make roller coster:\n',materials)
```

```python
labels = materials.values
sizes = [46, 24, 10, 10,10]
explode = (0.1, 0.1, 0, 0, 0)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.tight_layout()
plt.title('Most common types of materials used to make roller coaster')
plt.legend(materials.index, bbox_to_anchor = (1,0.5))
plt.show()
```
